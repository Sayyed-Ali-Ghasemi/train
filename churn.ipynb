```python
Q1 = df['MonthlyCharges'].quantile(.25)
Q3 = df['MonthlyCharges'].quantile(.75)

IQR = Q3 - Q1

U = df['MonthlyCharges'] > Q3 + (1.5*IQR)
L = df['MonthlyCharges'] < Q1 - (1.5*IQR)

outlier_U = df.loc[U, 'MonthlyCharges']
outlier_L = df.loc[L, 'MonthlyCharges']
```

```python
print(len(outlier_U),len(outlier_L))
#بررسی باکس پلات برای ستونی که احتمال دارد داده پرت داشته باشد
```

```python
x = df.drop("Churn", axis =1)
y = df["Churn"]

#جدا سازی هدف و بیقه اطلاعات
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, Ytrain, Ytest = train_test_split(x , y , test_size = 0.25, random_state=32)

print(Xtrain.shape)
print(Xtest.shape)

print(Ytrain.shape)
print(Ytest.shape)

#جدا سازی داده های تست و یادگیری
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()

DT.fit(Xtrain, Ytrain)
#روی چه دیتایی فیت بشه

pred = DT.predict(Xtest) #validation


#یاد دادن
```

```python
(pred == Ytest).mean()
#
```

```python
from sklearn.model_selection import GridSearchCV
#زدن گرید سرچ برای به دست آوردن بهترین حالت در درخت تصمیم گیری

DT = DecisionTreeClassifier()

param = {'criterion':['gini', 'entropy'] ,
         'max_depth':[3,4,5,6,7,8] ,
         'min_samples_split':[3,4,5,6] ,
         'min_samples_leaf':[2,3,4],
         'random_state':[32,23,17,13]}

GS = GridSearchCV(DT, param ,cv = 10, scoring = "f1", )
#داده های بالانس نداریم

GS.fit(x, y)


GS.best_score_
```

```python

GS.best_params_
#این گریدی سرچ بهترین نتیجش رو بهم بده
#بهترین نتیجه پارامتر هارو بهت میده
```

```python
DT = DecisionTreeClassifier(criterion = "entropy", max_depth = 6,min_samples_split = 3, min_samples_leaf = 4, random_state = 32)
```

```python
DT.fit(Xtrain, Ytrain)
#روی چه دیتایی فیت بشه

pred = DT.predict(Xtest)

(pred == Ytest).mean()
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
#Metrics:
acc = accuracy_score(Ytest, pred)
#برای وقتی هست که داده ها متوازن باشن
#مثل دیتاست گل ها که همگی متوازن بودن

rec = recall_score(Ytest, pred)

pre = precision_score(Ytest, pred)

fm = f1_score(Ytest, pred)

conf = confusion_matrix(Ytest, pred)

# tn - fp
# fn - tp

print(acc, rec, pre, fm)
print(conf)
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score

DT = DecisionTreeClassifier(criterion = "entropy", max_depth = 6,min_samples_split = 3,min_samples_leaf = 4,
                            class_weight = {0:0.3 , 1:0.7}, random_state =32)

acc = cross_val_score(DT, x , y , cv = 10, scoring = 'accuracy')
print(acc)
print("------------------------------------------------")
rec = cross_val_score(DT, x , y , cv = 10, scoring = 'recall')
print(rec)
print("------------------------------------------------")
pre = cross_val_score(DT, x , y , cv = 10, scoring = 'precision')
print(pre)
print("------------------------------------------------")
fm = cross_val_score(DT, x , y , cv = 10, scoring = 'f1')
print(fm)
print("------------------------------------------------")

print(acc.mean(), rec.mean(), pre.mean(), fm.mean())
#استفاده از کراس ولیدیشن
```

```python
from sklearn.naive_bayes import BernoulliNB, GaussianNB , MultinomialNB

#نایو بیز توزیع گائوسین
GNB = GaussianNB()
GNB_scores = cross_val_score(GNB, x , y , cv =10 , scoring = 'accuracy')
print(GNB_scores.mean())
print("-------------------------------")
#نایو بیز توزیع چند جمله ای
MNB = MultinomialNB()
MNB_scores = cross_val_score(MNB, x , y , cv =10 , scoring = 'accuracy')
print(MNB_scores.mean())
print("-------------------------------")
#نایو بیز توزیع برنولی
BNB = BernoulliNB()
BNB_scores = cross_val_score(BNB, x , y , cv =10 , scoring = 'accuracy')
print(BNB_scores.mean())
```
